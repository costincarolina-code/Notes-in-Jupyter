## Random numbers
#### Documentation for random number generators with numpy available at https://numpy.org/devdocs/reference/random/index.html. For non-uniform pdf look at https://numpy.org/devdocs/reference/random/legacy.html

```python
import numpy as np
import secrets # This contains functions to generate large random seeds
```

### Generation of uniform r.v.'s 

```python
# Set up initial seed
seed = secrets.randbits(128)   # Generate large positive integer: best kind of seed for the generator
print('seed = ', seed, '\n')

# Set up object rng = uniform random number generator (with default numpy choice = PCG-64 generator)
rng = np.random.default_rng(seed)

# Now we simply use the function 'random', applied to the object rng, to generate uniform istributed r.v.'s
# We can arrange them in matrices with desired shape.

# Just one variable
x = rng.random()
print('x = ', x, '\n')

# 100 r.v.'s arranged in a vector 
rvs = rng.random(100)
print('Vector = ', rvs, '\n')

# 100 r.v.'s arranged in a 10x10 matrix
rvs2 = rng.random( (10,10) )
print('Matrix = ', rvs2, '\n')
```

```python
# Generating Gaussian r.v.'s and plotting histogram

import numpy as np
import secrets
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation
mean = 3.5
std = 8.2
# Size of sample
N = 100000

# Seed
seed = secrets.randbits(32)
np.random.seed(seed)
#print(seed)
# Generating data
rvs = np.random.normal(mean,std,size=N)

# Generating pdf to overplot
xmin = mean - 5.*std
xmax = mean  + 5.*std
step = 0.1*std
x_axis = np.arange(xmin,xmax,step)
pdf = norm.pdf(x_axis,mean,std)

# Data histogram
plt.hist(rvs, bins=100, density=True, ec = 'black', histtype='bar')
# pdf plot
plt.plot(x_axis,pdf,label='$\mu$ = ' + str(mean) + ',' + ' $\sigma$ = ' + str(std))
plt.legend()
plt.xlabel('x',size = 18)
plt.ylabel('$N(\mu,\sigma)$', size = 18)
plt.show()
```

### Exercise from slides: simulating a star cluster

#### Class "cluster". Contains: 
- Definition of cluster object, with its relevant properties
- Functions to assign masses, positions, velocities (see slides for details)
- Functions to compute kinetic and gravitational potential energies and to check viral equilibrium
- Functions to make plots 

```python
import secrets
import numpy as np
import scipy.stats as sps
from matplotlib import pyplot as plt


def pol2cart(r,theta,phi):
    # Converting polar into Cartesian coordinates    
    x = r*np.sin(theta)*np.cos(phi)
    y = r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(theta)
    return x,y,z    


class cluster:
     
    def __init__(self,N,M,x,v,s,r,m1,m2):
        self.size = N  # Number of stars
        self.masses = M  
        self.positions = x
        self.velocities = v
        self.sigma = s
        self.radius = r
        self.mmin = m1
        self.mmax = m2

    
    def empty_cluster():
        N = 0
        M = 0.
        x = 0.
        v = 0.
        s = 0.
        r = 0.
        m1 = 0.
        m2 = 0.
        return cluster(N,M,x,v,s,r,m1,m2)

    
    def build(size=1000,sigma=0.5,radius=1.,mmin=0.1, mmax=150.):
        clust = cluster.empty_cluster()
        clust.assign_nstars(size)
        clust.assign_sigma(sigma)
        clust.assign_radius(radius)
        clust.assign_mmin(mmin)
        clust.assign_mmax(mmax)
        clust.assign_masses()
        clust.assign_velocities()
        clust.assign_positions()
        return clust

    
    def assign_nstars(self,N):
        self.size=N
  
    def assign_radius(self,r):
        self.radius = r
      
    def assign_sigma(self,sigma):
        self.sigma = sigma

    def assign_mmin(self,m):
        self.mmin = m
    
    def assign_mmax(self,m):
        self.mmax = m
   
 
    def assign_masses(self):
        # Assign masses assuming Salpeter distribution with mmin = 0.1 and mmax = 150 solar masses
        seed = secrets.randbits(128)
        rng = np.random.default_rng(seed)
        x = rng.random(self.size)  
        # After generating uniform variables, transform to Salpeter distributed r.v.'s
        mmin = self.mmin
        mmax = self.mmax
        alpha = 2.3
        t = 1./(1.-alpha)
        mstars = ( (mmax**(1.-alpha) - mmin**(1.-alpha))*x + mmin**(1.-alpha) )**t
        self.masses = np.array(mstars)
    
        
    def assign_velocities(self):
        seed = secrets.randbits(32)
        np.random.seed(seed)
        rng = np.random.default_rng(seed)
        # Magnitudes of stars' velocities fom Maxwellian distribution
        vmag = sps.maxwell.rvs(loc=0, scale=self.sigma, size=self.size)
        # Velocity orientation, polar angle
        vtheta = rng.random(self.size)
        vtheta = np.arccos(1.-2.*vtheta)
        # Velocity orientation, azimuthal angle
        # Azimuthal angles of stars in the cluster
        vphi = rng.random(self.size)
        vphi = 2.*np.pi*vphi
        # Transform polar -> cartesian
        vx,vy,vz = pol2cart(vmag,vtheta,vphi)
        # Save cartesian coordinates in Nx3 array (each row corresponds to vx,vy,vz of a given star)
        vstars = np.transpose(np.array( [ vx,vy,vz ], dtype=float))
        self.velocities = vstars
    
    
    def assign_positions(self):
        # Assign positions assuming Plummer density 
        seed = secrets.randbits(128) 
        rng = np.random.default_rng(seed)
        # Radial coordinates of stars in the cluster
        r = rng.random(self.size)
        a = self.radius
        t = -2./3.
        r = np.sqrt(a**2/(r**t-1.0))
        # Polar angles of stars in the clusters
        theta = rng.random(self.size)
        theta = np.arccos(1.-2.*theta)
        # Azimuthal angles of stars in the cluster
        phi = rng.random(self.size)
        phi = 2.*np.pi*phi
        # Transform polar -> cartesian
        x,y,z = pol2cart(r,theta,phi)
        # Save cartesian coordinates in Nx3 array (each row corresponds to x,y,z of a given star)
        xstars = np.transpose(np.array( [ x,y,z ], dtype=float))
        self.positions = xstars


    def gravitational_energy(self):
        # Calculates the total gravitational energy
        # of the cluster
        G = 6.6743e-11  # S.I. units
        pc = 3.086e16   # 1 pc in m
        Msun = 1.989e30 # solar mass in kg
        # Cartesian distances between 1 star an all the others 
        def dist(x1,x2):
            d = np.sqrt( (x1[0] - x2[:,0])**2 + (x1[1] - x2[:,1])**2 + (x1[2] - x2[:,2])**2 )  
            return d # This is a vector of the same length as x2
            # Computes gravitational energy
        E = 0.
        for i in range(self.size-1):
            # Position of i-th star in the list
            x1 = self.positions[i]
            # Mass of i-th star (in kg)
            m1 = Msun*self.masses[i]
            # Positions of all stars labeled by j, with j>i in the list
            x2 = self.positions[i+1:]
            # Masses of all j>i stars (in kg)
            m2 = Msun*self.masses[i+1:]
            # Distances between all i,j pairs (in m)
            distances = pc*dist(x1,x2)
            # Gets energies of all pairs (in N m^2)
            Gm1m2 = G*m1*m2
            energy = Gm1m2*distances**(-1)
            # Sums and add to total potential energy
            E -= np.sum(energy)
        return E

    
    def kinetic_energy(self):
        # Calculates total kinetic energy of the cluster
        Msun = 1.989e+30      # Solar mass in kg
        km = 1e+3             # km in m        
        # Computes v^2 for all particles in the cluster and convert km/s in m/s
        v2 = (self.velocities[:,0])**2 + (self.velocities[:,1]**2) + (self.velocities[:,2])**2
        v2 = km*km*v2
        # Computes total kinetic energy (in N m^2)
        E = 0.5*np.sum(Msun*self.masses*v2)
        return E
    
    
    def check_virialization(self):
        K = self.kinetic_energy()
        W = self.gravitational_energy()
        Qvir = 2.*K/(abs(W))
        return Qvir
    
    def virialize(self):
        Qvir=self.check_virialization()
        self.velocities = self.velocities/np.sqrt(Qvir)
        
        
    def display_cluster_projection(self,x1,x2,distance):
        x = clust.positions[:,x1]
        y = clust.positions[:,x2]
        coord = ['x','y','z']
        xmax = distance*self.radius
        xmin = -xmax
        plt.axis('square')
        plt.xlabel(coord[x1] + ' (pc)',size=16)
        plt.ylabel(coord[x2] + ' (pc)',size=16)
        plt.xlim(xmin,xmax)
        plt.ylim(xmin,xmax)
        plt.scatter(x,y, s=2, color='blue')
        plt.show()
    
    
    def display_pdf(self,kind,nbins):
        if (str(kind) == 'velocity'):
            x = (self.velocities[:,0])**2 + self.velocities[:,1]**2 + self.velocities[:,2]**2
            plt.xlabel('v (km/s)', size = 16)
        elif (str(kind) == 'radius'):
            x = np.sqrt((self.positions[:,0])**2 + self.positions[:,1]**2 + self.positions[:,2]**2)
            plt.yscale('log')
            plt.xlabel('r (pc)', size = 16)
            plt.xlim(1e-1*self.radius,1e2*self.radius)
        plt.xscale('log') 
        plt.ylabel('PDF', size = 16)
        plt.hist(x, bins=nbins, density=True, histtype='step', linewidth=3)
        plt.show()
        
    
    def display_mass_distribution(self):
        # Find radial distances from the center of stars in the cluster
        r = np.array(np.sqrt((self.positions[:,0])**2 + self.positions[:,1]**2 + self.positions[:,2]**2))      
        # Find indeces of r that correspond to sorting distances from smallest to larges
        ind = r.argsort()
        x = r[ind]     
        # Rearranges masses, ordering M(r) for r from smallest to largest
        M = self.masses[ind]
        Mr = [ ]
        #Gets total mass in spheres of growing radius r
        for i in range(self.size):
            Mr.append(np.sum(M[0:i]))
        plt.plot(x,Mr)
        plt.xlabel('r (pc)', size=16)
        plt.ylabel('M(r) $\, M_{\odot}$', size=16)
        plt.show()
```

#### Simulate cluster

```python
# Input cluster specs
N = 10000   # Total number of stars in the cluster
s = 0.5     # Velocity dispersion (scale parameter in Maxwell distribution, used to generate velocities)
r = 1.      # Cluster "radius"  (parameter in Plummer distribution, used to generate positions)
mmin = 0.1
mmax = 150

# Generates star positions, masses and velocities
clust = cluster.build(radius=r,sigma=s,size=N, mmin = mmin, mmax = mmax)
```

#### Calculates cluster energy and virialize

```python
W = clust.gravitational_energy()
print('Total gravitational energy = ', W)
K = clust.kinetic_energy()
print('Total kinetic energy = ', K)
Qvir = clust.check_virialization()
print('Qvir = ', Qvir)
clust.virialize()
Qvir = clust.check_virialization()
print('Qvir after normalizing velocities = ', Qvir)
```
