```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2D PLOTS
```

```python
# Input functions to plot
twopi = 2*np.pi
step = 0.05*twopi
x = np.arange(0,twopi,step)
y = np.cos(x)
z = np.sin(x)
```

```python
# Basic plot
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# Embellishing

#Changing line style, lineweight and colors
cos=plt.plot(x,y,label='cos(x)',linestyle='dashed',linewidth=2,color='red')
sin=plt.plot(x,z,label='sin(x)',linestyle='dashdot',linewidth=2,color='green')

# Adding labels on x-y axis
plt.xlabel('x',fontsize = 16) 
plt.ylabel('y', fontsize = 16)
# Adding legend (after specifying "label=" in plt.plot above)
plt.legend(loc='lower left', fontsize = 14)

plt.show()
```

```python
# Scatter plot

plt.scatter(x,y,label='cos(x)', marker = '^',color='blue')
plt.scatter(x,z,label='sin(x)', marker = 'o',color='gray')

# Change axis boundaries
plt.xlim(-0.3,6.2)
plt.ylim(-1.2,1.2)

# Adding labels on x-y axis
plt.xlabel('x',fontsize = 16) 
plt.ylabel('y', fontsize = 16)
# Adding legend (after specifying "label=" in plt.plot above)
plt.legend(loc='lower left', fontsize = 14)

plt.show()
```

```python
# Points and lines

plt.plot(x,y,label='cos(x)',linestyle='dashed',linewidth=1.5,color='lightsteelblue')
plt.plot(x,z,label='sin(x)',linestyle='dashdot',linewidth=1.5,color='gray')

# List of marker sizes
size = [64 for i in range(len(x))]
  
plt.scatter(x,y,label='cos(x)', marker = '^',color='blue',s=size)
plt.scatter(x,z,label='sin(x)', marker = 'o',color='black',s=size)

# Change axis boundaries
plt.xlim(-0.3,6.2)
plt.ylim(-1.2,1.2)

# Adding labels on x,y-axis
plt.xlabel('x',fontsize = 16) 
plt.ylabel('y', fontsize = 16)

# Change size of ticks on x and y-axis
plt.xticks(size=14)
plt.yticks(size=14)

# Adding legend (after specifying "label=" in plt.plot above)
plt.legend(loc='lower left', fontsize = 13)

# Plot title
plt.title('Sine and cosine', size = 18)
plt.savefig('2dplot_example')

plt.show()
```

```python
# Panel with 2 plots

# Uses subplot and defines 2 plots with separate axes
fig, axs = plt.subplots(2)
axs[0].plot(x,y)
axs[1].plot(x,z)
```

```python
# Horizontally stacked subplots
fig, axs = plt.subplots(1,2,figsize=(9,3))
axs[0].plot(x,y,label='cos(x)')
axs[1].plot(x,z,label='sin(x)')

axs[0].set_xlabel('x',fontsize = 16)
axs[1].set_xlabel('x',fontsize=16)
axs[0].label_outer()
axs[1].label_outer()

axs[0].legend(fontsize = 13)
axs[1].legend(fontsize = 13)
```

```python
# Array of 4 subplots

twopi = 2*np.pi
step = 0.05*twopi

# Defining functions
x = np.arange(0,twopi,step)
y = np.cos(x)
z = np.sin(x)

x1 = np.arange(-10,10)
t = np.arctan(x1)
s = np.exp(-x1)

# Array of subplots
fig, axs = plt.subplots(2,2,figsize=(7,4))

# Plots in each panel
axs[0,0].plot(x,y,label='cos(x)', color='red')
axs[1,0].plot(x,z,label='sin(x)',color='blue')
axs[0,1].plot(x1,t,label='arctan(x)',color='green')
axs[1,1].plot(x1,s,label='log(x)',color='cyan')

# x labels on bottom plots
axs[1,0].set_xlabel('x',fontsize = 16)
axs[1,1].set_xlabel('x',fontsize = 16)
axs[0,0].set_ylabel('y',fontsize = 16)
axs[1,0].set_ylabel('y',fontsize = 16)

# Setting logscale on bottom left panel
axs[1,1].set_yscale('log')

axs[0,0].legend(fontsize = 9, loc='upper center')
axs[1,0].legend(fontsize = 9)
axs[0,1].legend(fontsize = 9, loc='upper left')
axs[1,1].legend(fontsize = 9)

axs[0,0].label_outer()
# Removes tick labels from upper right panel
axs[0,1].set_xticklabels([])

# Overall title on top
fig.suptitle("Plotting some functions")
plt.show()
```

```python
# Creating a 2d grid

# This creates 10 linearly spaced points in the interval [-5,5]
x = np.linspace(-5,5,10)
# Same in y
y = np.linspace(-5,5,10)

# This creates a coordinate grid
X,Y = np.meshgrid(x,y)
# Let's check how it looks
plt.scatter(X,Y,marker='o')
plt.show()
```

```python
# Now we evaluate some functions on a grid and build contour plots

# This creates 100 linearly spaced points in the interval [-5,5]
x = np.linspace(-5,5,100)
# Same in y
y = np.linspace(-5,5,100)

# This creates a coordinate grid
X,Y = np.meshgrid(x,y)
# Evaluate the function x^2 - y^2 on the grid (vectorized version)
Z = X**2 - Y**2

plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
# Note the Latex style syntax if you want to type formulas
plt.title('Level contours of $Z = X^2 - Y^2$')

# Plotting contour, '10' is the number of contour lines we ask to plot)
curves = plt.contour(X,Y,Z,10)
# Writes levels on the plot
plt.clabel(curves)
```

```python
# Sme contour plot as above but I am not using vectorization after meshgrid

# This creates 100 linearly spaced points in the interval [-5,5]
x = np.linspace(-5,5,100)
# Same in y
y = np.linspace(-5,5,100)

# This creates a coordinate grid
X,Y = np.meshgrid(x,y)

###################################################
# This part is different from box above

# Compute x^2 + y^2 as scalar function in all the (x,y) points 
# and store in z coordinate
z = [ ]
for i in x:
    for j in y:
        z.append(i**2 - j**2)

# Now turn z from list into array and reshape it to make
# it become a 2-dim array with 1 to 1 correspondence with the 
# x,y grid
z = np.array(z)
Z = z.reshape(len(y),len(x))

# The rest is as in the previous box...
#########################################################

plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
# Note the Latex style syntax if you want to type formulas
plt.title('Level contours of $Z = X^2 - Y^2$')

# Plotting contour, '10' is the number of contour lines we ask to plot)
curves = plt.contour(X,Y,Z,10)
# Writes levels on the plot
plt.clabel(curves)
plt.show()
```

```python
# Filled contours
curvesf = plt.contourf(X,Y,Z,10)

plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
# Note the Latex style syntax if you want to type formulas
plt.title('Filled level contours of $Z = X^2 - Y^2$')

plt.clabel(curvesf,colors='black')
#plt.colorbar()
```

```python
# Density plot
curvesf = plt.pcolormesh(X,Y,Z,cmap='gray')

plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
# Note the Latex style syntax if you want to type formulas
plt.title('Density plot of $Z = X^2 - Y^2$')

#plt.clabel(curvesf,colors='black')
plt.colorbar()
plt.show()
```

```python
# Density plot using imshow

from pylab import imshow, show
#imshow(Z,cmap='gray')

imshow(Z,cmap='gray',extent=[-5,5,-5,5])

plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
# Note the Latex style syntax if you want to type formulas
plt.title('Density plot of $Z = X^2 - Y^2$')

plt.colorbar()
plt.show()
```

```python
# Multivariate Gaussian contours

from scipy.stats import multivariate_normal as multiv
from scipy.linalg import det

mean = [3.,3.]
cov = [ [1.3, 0.7], [0.7, 2.1] ] 
# This produces a bivariate Gaussian function with mean and covariance given right above
bivariate = multiv(mean, cov) 

x,y = np.mgrid[-6:12:3600j, -6:12:3600j]
# Creates al possible (x,y) pairs to pass as coordinates to bivariate
pos = np.dstack( (x,y) )

#print(pos)

# Evaluate bivariate in all x,y points
z = bivariate.pdf(pos)
#print(z)


# Generatres X,Y grid for plotting
#X,Y = np.meshgrid(x,y)

contours=plt.contourf(x,y,z,levels=[0.000007,0.01,0.1])
plt.colorbar()
```

```python
# Importing maxwell class from scipy.stats
from scipy.stats import maxwell

# Distribution parameters
kB = 1.380649*1e-23
T = 3000
u = 1.661*1e-27
m = 4.*u

# Number of data points to generate
N = 100000
# Generating N Maxwell distributed random variables and storing in vector data
data = maxwell.rvs(loc=0, scale=np.sqrt(kB*T/m), size=N)

print(data[1:100])

# Producing and plotting a histogram of data with 100 bins; density = True means that we want
# to normalize by dividing each column by the total number of counts
plt.hist(data, bins=100, density=True, alpha=0.3, ec = 'black', histtype='bar')

x = np.arange(0,np.max(data))
# Generating and plotting the Maxwell pdf, for comparison with data histogram
y = maxwell.pdf(x, loc=0, scale=np.sqrt(kB*T/m))

#print(len(ticks))
#print(ticks)

plt.xlabel(r'$v \, ({m/s})$', fontsize = 16)
plt.ylabel('Frequency', fontsize = 16)

plt.plot(x,y,label='Helium, T = ' + str(T) + ' K')
plt.legend(fontsize = 14)
plt.show()
```
