### How to define vavriables in Python

```python
a = 1 # now a is an integer number
print(type(a))
#if we add a rational number it becomes a float number (a number wth the comma)
print(type(a+0.5))
print(type(10), type(10.0))
```

```python
# the difference between integers and float numbers
# is important because some functions require integer numbers as input,
# like the range function
# if we do a for cycle like the following
for i in range(10):
    print(i)
# we will obtain the right result
# but if we run it with a float, we will get an error
for i in range(10.):
    print(i)
```

```python
# we can also define strings, which have to be defined using '' or ""
b = 'ciao1!'
print(type(b))
```

```python
# we can sum only variables of the same type
b + 1.0
```

```python
c = b + ' Marco'
print(c)
```

# defining functions

```python
def func(x):
    a1 = 1
    b1 = 2
    c1 = 0.5
    return a1*x**2 + b1*x + c1
```

```python
func(0.1)
```

```python
func(xs)
```

```python
def func_smart(x, a1=1., b1=2., c1=0.5):
    '''
    Returns a parabola, given a1, b1, c1
    
    Input
    x: array_like, points that you want to plot
    a1: factor of x^2 (default: 1)
    b1: factor of x (default: 2)
    c1: constant factor (default: 0.5)
    
    Output
    returns a1*x^2 + b1*x + c1
    '''
    return a1*x**2 + b1*x + c1
```

```python
func_smart(xs, a1 = 2, b1 = 0.)
```

```python
plt.figure(figsize = (10,7))
plt.plot(xs, func_smart(xs), label = 'a1=1, b1=2, c1=0.5', color='blue')
plt.plot(xs, func_smart(xs, a1=0.5), label = 'a1=0.5, b1=2, c1=0.5', color = 'red')
plt.plot(xs, func_smart(xs, b1=3), label = 'a1=1, b1=3, c1=0.5', color = 'darkorange')

plt.xlabel('x', fontsize = 20)
plt.ylabel('y(x)', fontsize = 20)

plt.legend(fontsize = 20)
```
